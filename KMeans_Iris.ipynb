```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
from sklearn.datasets import load_iris

iris_data = load_iris()
iris = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)

#add target column
iris['Species'] = iris_data.target
iris['Species'] = iris['Species'].map({0: 'Iris-setosa', 1: 'Iris-versicolor', 2: 'Iris-virginica'})
iris.head()
```

```python
#show data by species in form of pie chart
fig = px.pie(iris, names='Species', title='Data Distribution')
fig.show()

#allow visualizations by using different diagrams
fig = px.box(iris, x='Species', y='sepal length (cm)', color='Species', title='Sepal Length Distribution')
fig.show()

fig = px.histogram(iris, x='sepal length (cm)', color='Species', title='Sepal Length Histogram', nbins=30)
fig.show()

fig = px.scatter(iris, x='sepal length (cm)', y='sepal width (cm)', color='Species', title='Sepal Length vs Sepal Width')
fig.show()
```

```python
X = iris_data.data
y = iris['Species']  # this is only for visualization i havent used it in kmeans

# find the no, of clusters that best suit our data
sse = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, max_iter=300)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)

fig = px.line(x=range(1, 11), y=sse, title='Optimal Clusters', labels={'x': 'Number of Clusters', 'y': 'SSE'})
fig.show()

kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
iris['Cluster'] = kmeans.fit_predict(X)

#show the clustering result
fig = px.scatter(iris, x='sepal length (cm)', y='sepal width (cm)', color='Cluster', title='K-Means Clustering Results')
fig.show()
```

```python
fig = go.Figure()

#add scatter plot for every cluster
for cluster in range(3):
    fig.add_trace(go.Scatter(
        x=iris[iris['Cluster'] == cluster]['sepal length (cm)'],
        y=iris[iris['Cluster'] == cluster]['sepal width (cm)'],
        mode='markers',
        marker_color=['#DB4CB2', '#c9e9f6', '#7D3AC1'][cluster],
        name=f'Cluster {cluster + 1}'
    ))

#show centroids
fig.add_trace(go.Scatter(
    x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1],
    mode='markers', marker_color='black', marker_symbol='x', marker_size=10,
    name='Centroids'
))

fig.update_layout(title='K-Means Clustering with Centroids', width=800, height=600)
fig.show()
```
